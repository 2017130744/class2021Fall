```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd

def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0

A  = 440.00
Bb = 466.16
B  = 493.88
C  = 523.25
Dd = 554.37
D  = 587.33
Eb = 622.25 
E  = 659.25	 
F  = 698.46
Gb = 739.99 
G  = 783.99
Ab = 830.61

s_all = np.empty(0)
```

```python
# The bridge on the river Kwai
# 16음표 0.15, 8분음표 0.3, 4분음표 0.6

notes = [C/2, G/2, D, E, 0, E/2, B/2, C, G, 0
         F/2, B/2, C, A, 0, G/2, D, G, B, Ab/2, E, Ab, B]

durs = [0.15,0.15,0.15,0.15,0.6,0.15,0.15,0.15,0.15,0.6
        ]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```
